```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Calculate correlation coefficient
correlation = df_combined['VL'].corr(df_combined['AUM (€)'])
# Calculate p-value
slope, intercept, r_value, p_value, std_err = stats.linregress(df_combined['VL'], df_combined['AUM (€)'])

# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# Plot 1: Time series of both variables
ax1.plot(df_combined['Date'], df_combined['VL'], label='VL', color='blue')
ax2_twin = ax1.twinx()  # Create a twin axis
ax2_twin.plot(df_combined['Date'], df_combined['AUM (€)'], label='AUM (€)', color='red')

# Add labels and title for first plot
ax1.set_xlabel('Date')
ax1.set_ylabel('VL', color='blue')
ax2_twin.set_ylabel('AUM (€)', color='red')
ax1.set_title('Time Series Comparison')

# Rotate x-axis labels
ax1.tick_params(axis='x', rotation=45)

# Add legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2_twin.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

# Plot 2: Scatter plot with regression line
ax2.scatter(df_combined['VL'], df_combined['AUM (€)'], alpha=0.5)
# Add regression line
x = df_combined['VL']
y = df_combined['AUM (€)']
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
ax2.plot(x, p(x), "r--", alpha=0.8)

# Add correlation information
ax2.text(0.05, 0.95, f'Correlation: {correlation:.2f}\np-value: {p_value:.4f}',
         transform=ax2.transAxes,
         bbox=dict(facecolor='white', alpha=0.8))

# Add labels and title for second plot
ax2.set_xlabel('VL')
ax2.set_ylabel('AUM')
ax2.set_title('Correlation Analysis')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()

# Print statistics
print("\nCorrelation Analysis:")
print(f"Correlation coefficient: {correlation:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"R-squared: {r_value**2:.4f}")

# Basic statistics for both series
print("\nBasic Statistics:")
print("\nVL Statistics:")
print(df_combined['VL'].describe())
print("\nAUM Statistics:")
print(df_combined['AUM (€)'].describe())
```

```python
def extract_values_by_dates(source_df, dates_list, value_column=None, date_column=None):
    # Convert dates_list to datetime if they aren't already
    dates_list = pd.to_datetime(dates_list)

    # Handle DataFrame date column/index
    if date_column is not None:
        source_df = source_df.copy()
        source_df[date_column] = pd.to_datetime(source_df[date_column])
        source_df.set_index(date_column, inplace=True)
    else:
        # Ensure index is datetime
        source_df.index = pd.to_datetime(source_df.index)

    # If no value column specified and only one column exists, use that
    if value_column is None:
        if len(source_df.columns) == 1:
            value_column = source_df.columns[0]
        else:
            raise ValueError("Multiple columns found. Please specify value_column.")

    # Extract values for the specified dates
    try:
        result = source_df.loc[dates_list, value_column]

        # Check for missing dates
        missing_dates = dates_list[~dates_list.isin(source_df.index)]
        if len(missing_dates) > 0:
            print(f"Warning: No data found for dates: {missing_dates.tolist()}")

        return result

    except KeyError as e:
        raise KeyError(f"Column '{value_column}' not found in DataFrame") from e

perf_fund.head()
```
